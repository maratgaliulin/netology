```python
import requests
```

```python
def expensive_currency():                               
        currency_dict = requests.get('https://www.cbr-xml-daily.ru/daily_json.js').json()['Valute']
        expensive_value = 0
        expensive_value_name = ''
        for key in currency_dict:
            if currency_dict[key]['Value'] > expensive_value:
                expensive_value = currency_dict[key]['Value']
                expensive_value_name = currency_dict[key]['Name']
        print(f"Валюта с максимальным значением курса: {expensive_value_name}, {expensive_value}")
```

```python
# параметр diff добавлен (на него указывает комментарий)

class Rate:
    def __init__(self, format_='value', diff=False):
        self.format = format_
        self.diff = diff

    def __exchange_rates(self):                           # для эксперимента сделал метод exchange_rates приватным, и убедился - работает) (по крайней мере в PyCharm) -  вне класса доступ к методу пропадает
        """
        Возвращает ответ сервиса с информацией о валютах
        """

        self.r = requests.get('https://www.cbr-xml-daily.ru/daily_json.js')
        return self.r.json()['Valute']

    def make_format(self, currency):
        response = self.__exchange_rates()
        if currency in response:
            if self.format == 'full':
                return response[currency]
            if self.format == 'value':
                if not self.diff:                            # условие для параметра diff: если true, возвращает разницу между 'Value' и 'Previous', если False, 
                    return response[currency]['Value']       # то возвращает 'Value'
                else:
                    return round((response[currency]['Value'] - response[currency]['Previous']), 2)
        return 'Error'

    def eur(self, diff):                                     # указывается параметр diff, если он равен true, возвращается разность курсов
        """Возвращает курс евро на сегодня в формате self.format"""
        self.diff = diff
        return self.make_format('EUR')

    def usd(self, diff):
        """Возвращает курс доллара на сегодня в формате self.format"""
        self.diff = diff
        return self.make_format('USD')

    def brl(self):
        """Возвращает курс бразильского реала на сегодня в формате self.format"""
        return self.make_format('BRL')

    
```

```python
rate_instance = Rate()
rate_instance.usd(diff=True)
```

Напишите класс Designer, который учитывает количество международных премий. Подсказки в коде занятия в разделе “Домашнее задание задача 3”.

```python
class Employee:
    def __init__(self, name, seniority):
        self.name = name
        self.seniority = seniority
        self.grade = 1

    def grade_up(self):
        """Повышает уровень сотрудника"""
        self.grade +=1

    def publish_grade(self) -> object:
        """Публикация результатов аккредитации сотрудников"""
        print(self.name, self.grade)

    def check_if_it_is_time_to_upgrade(self):
        pass


class Designer(Employee):
    def __init__(self, name, seniority):
        super().__init__(name, seniority)



    def check_if_it_is_time_to_upgrade(self, award = False):
        # Для каждой аккредитации увеличиваем счетчик на 1 балл
        # Если получил международную премию (award = True), то добавляем 2 балла
        if not award:
            self.seniority += 1
        elif award:
            self.seniority += 2

        # Условие повышения сотрудника из презентации - условие с остатком деления на 7 не работает как в случае с Developer, т.к. если дизайнеру сразу прибавить количество баллов
        # до более 7 (например 8 или 9), то значение 7 перешагнется и grade-up не будет произведен. Поэтому решил обновлять информацию о grade 
        # с каждым вызовом функции check_if_it_is_time_to_upgrade() как целое число от деления количества баллов на 7, плюс исходный 1 grade. Вроде работает)
        self.grade = int(self.seniority / 7) + 1

        # Публикация результатов
        return self.publish_grade()
```

```python
des = Designer('Carren', 0)

for i in range(20):
  des.check_if_it_is_time_to_upgrade(award=True)

for b in range(20):
  des.check_if_it_is_time_to_upgrade(award=False)
```
