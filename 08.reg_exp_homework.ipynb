Напишите функцию, которая принимает на вход строку и проверяет является
ли она валидным транспортным номером (1 буква, 3 цифры, 2 буквы, 2-3 цифры).
Обратите внимание, что не все буквы кириллического алфавита используются в транспортных номерах.
Если номер валиден, то функция должна возвращать отдельно номер и регион.


```python
def car_number_identifier():
    car_number = input("Введите номер машины: ")
    pattern = r'([A-Z]{1})(\d{3})([A-Z]{2})(\d{2,3})'
    region_no = ''
    if re.match(pattern, car_number):
        region_no = re.split(r'([A-Z]{1})(\d{3})([A-Z]{2})', car_number)[-1]
        print(f"Номер {car_number} валиден. Регион: {region_no}")
    else:
        print("Номер не валиден")
```

```python
car_number_identifier()
```

Напишите функцию, которая будет удалять все последовательные повторы
слов из заданной строки при помощи регулярных выражений.

```python
def repeating_word_deleter():
    string_to_purge = input("Введите текст для очистки от повторяющихся слов: ")
    print(re.sub(r'\b([^\W\d_]+)(\s+\1)+\b', r'\1', string_to_purge, flags=re.I))
```

```python
repeating_word_deleter()
```

Напишите функцию, которая будет возвращать акроним по переданной в нее строке со словами.

```python
def acronym_creator():
    some_words = input("Введите название: ")
    acronym = "".join(item[0].upper() for item in re.findall("\w+", some_words))
    print(acronym)
```

```python
acronym_creator()
```

Напишите функцию, которая будет принимать на вход список email-адресов и выводить их распределение по доменным зонам.

```python
def domain_counter(emails: list):
    def pattern_count(emails: list, indiv_pattern: str):
        pattern_template = f'(\w+)({indiv_pattern})'
        counter = 0
        for e in emails:
            if re.match(pattern_template, e):
                counter += 1
        print(f'{indiv_pattern}: {counter}')
    pattern_count(emails, '@gmail.com')
    pattern_count(emails, '@test.in')
    pattern_count(emails, '@ya.ru')
    pattern_count(emails, '@mail.ru')
```

```python
emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']
domain_counter(emails)
```
