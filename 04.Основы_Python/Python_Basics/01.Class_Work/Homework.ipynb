совмещение таблиц country и country_language

```python
df_merged = pd.read_sql(
"""
select 
  *
from  
  country_language clang  
join
  country ct
on
  ct.code = clang.country_code
where
    is_official = 'True'
""", 

'postgresql://netology:NetoSQL2019@84.201.153.170:19001/world-db')
df_merged.head(55) 
```

совмещение двух таблиц методами pandas + сортировка по названию страны

```python
df_country_lang = pd.read_sql(
"""
SELECT 
  *
FROM
  country_language 
""", 

'postgresql://netology:NetoSQL2019@84.201.153.170:19001/world-db')
df_country_lang.head(55)
```

```python
df_country = pd.read_sql(
"""
select 
  *
from
  country
""", 

'postgresql://netology:NetoSQL2019@84.201.153.170:19001/world-db')
```

```python
df_pandas_merged = pd.merge(
    df_country_lang.loc[df_country_lang['is_official'] == True],
    df_country,
    left_on='country_code',
    right_on='code',
    how='outer'
).sort_values(by=['name'])

df_pandas_merged.head(55)
```
