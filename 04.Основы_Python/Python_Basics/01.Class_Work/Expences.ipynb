```python
import pandas as pd

exp = pd.read_csv('expences.csv')[
    ['Отчетный месяц (reporting month)', 'Название операции', 'Сумма', 'Категория', 'Месяц и цифра']]

print(exp.groupby('Категория').sum('Сумма').head(10))
print(exp.groupby('Категория').sum('Сумма').describe().head(10))

print(exp.groupby('Месяц и цифра').sum('Сумма').head(10))
print(exp.groupby('Месяц и цифра').sum('Сумма').describe().head(10))
```
