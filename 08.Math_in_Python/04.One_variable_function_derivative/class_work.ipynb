```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sympy as sym
```

1. Задание и визуализация функций

```python
def function(x):
    return 1/(x - 1)
```

```python
x = np.arange(0, 2*(np.pi), 0.1)

x
```

```python
y = np.sin(x)
y
```

```python
sns.lineplot(x=x, y=y, label='1st function')
sns.lineplot(x=x, y=function(x), label='2nd function')
sns.set_theme('notebook')
plt.legend()
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Our function');
```

Производные функции

```python
def h_func(x):
    return np.sin (x + 1)**2 

def h_func_1(x):
    return 2 * np.sin(x + 1) * np.cos(x + 1)
```

```python
x3 = np.arange(1, 10, 0.1)

plt.plot(x3, h_func(x3), label='Функция')
plt.plot(x3, h_func_1(x3), label='Производная')
plt.legend()
plt.show()
```

```python
def sin_func(x):
    return np.sin(x)

def cos_func(x):
    return np.cos(x)


plt.plot(x3, sin_func(x3), label='Функция')
plt.plot(x3, cos_func(x3), label='Производная')
plt.legend()
plt.show()
```
