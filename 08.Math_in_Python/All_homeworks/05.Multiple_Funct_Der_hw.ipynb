1) Посчитать частные производные функций:

```python
x = sym.Symbol('x')
y = sym.Symbol('y')
```

```python
# 2 способа нахождения частных производных при помощи Sympy
```

```python
z1 = 2 * x**2 * y**3 * 1/x + y**2 * x + 7

z1prime_x = z1.diff(x)
z1prime_y = z1.diff(y)
```

2) Вычислить градиент функции z1 (см. выше) в точке [2,1]

```python
x_num, y_num = 1,2

def grad_xy(x, y):
  x_coord = 2 * y**3 + y**2
  y_coord = 6 * x * y**2 + 2 * x * y
  return [x_coord, y_coord]
```

3) Построить трехмерный график функции

```python
fig = plt.figure(figsize = (10, 10))
ax = fig.gca(projection = '3d')
step = .25


X = np.arange(-10, 10, step)
Y = np.arange(-10, 10, step)

X, Y = np.meshgrid(X, Y)

R = Y**2 - X**2 * Y + 2 * X * Y


surf = ax.plot_surface(X, Y, R, cmap=cm.coolwarm, linewidth=0, antialiased=True)

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

fig.colorbar(surf, shrink=0.5, aspect=5)

ax.azim = 10
ax.dist = 10
ax.elev = 30

plt.show()
```
