Задание 4

Дана функция с неизвестными параметрами a и b, вам также известны значения функции fx в точках x.

Найдите оптимальные параметры a, b, минимизирующие абсолютную ошибку error.

```python
def f(x, a, b):
    return np.e**a * np.sin(b*x) + x
```

```python
x = np.array([0.        , 0.26315789, 0.52631579, 0.78947368, 1.05263158,
              1.31578947, 1.57894737, 1.84210526, 2.10526316, 2.36842105,
              2.63157895, 2.89473684, 3.15789474, 3.42105263, 3.68421053,
              3.94736842, 4.21052632, 4.47368421, 4.73684211, 5.        ])

fx = np.array([  0.        , -12.01819092, -18.90968634, -17.68786571,
                -8.7529108 ,   4.27524517,  16.06801336,  21.81250213,
                19.22059845,   9.48411207,  -3.22273056, -13.48576488,
               -16.91096359, -11.95866834,  -0.58630088,  12.56873816,
                22.12489421,  24.20292139,  18.04522521,   6.33211092])
```

```python
plt.plot(x, fx)
plt.show()
```

```python
def error(params):
    return np.sum(np.abs(fx - f(x, params[0], params[1])))
```

```python
params_for_testing = np.array([])
```

```python
np.random.seed(20)

for i in range(20):
  p0 = np.random.uniform(low=-7, high=7, size=(2,))
  res = curve_fit(f, x, fx, p0 = p0, method = 'lm')
  print(p0)
  params_for_testing = np.append(params_for_testing, p0)
  plt.figure(figsize=(12, 8))
  plt.plot(x, fx, color='blue')
  plt.plot(x, f(x, res[0][0], res[0][1]), color='red')
  plt.show()
```

```python
params_for_testing = params_for_testing.reshape(20,2)
```

```python
least_error = params_for_testing[0]

for i in range(len(params_for_testing)):
  print(f"{params_for_testing[i]}, {error(params_for_testing[i])}")
  if((i+1) < len(params_for_testing)):
    if(error(params_for_testing[i+1])<error(params_for_testing[i])):
      least_error = params_for_testing[i+1]
```

```python
print(f"Наименьшее значение error с параметрами a,b: {least_error}")
```

Необходимо найти наиболее похожего пользователя по косинусной метрике для каждого из пользователей (ответ: 10 пар вида (x, y), где y - наиболее похожий пользователь на пользователя x).

```python
ratings
```

```python
def angle_bw_vectors(vec1, vec2):
    dot_vec_1_vec_2 = vec1.dot(vec2)
    norm_vec_1_vec_2 = np.linalg.norm(vec1) * np.linalg.norm(vec2)

    return np.rad2deg(np.arccos(dot_vec_1_vec_2 / norm_vec_1_vec_2))
```

```python
for i in range(len(ratings)):
  k_vect = [0, 0, 90]
  for k in range(len(ratings)):    
    if (i != k):
      # print(i, k)
      # print(angle_bw_vectors(ratings[i], ratings[k]))
      if(k_vect[2] > angle_bw_vectors(ratings[i], ratings[k])):
        k_vect = [i, k, angle_bw_vectors(ratings[i], ratings[k])]
      else: k_vect = k_vect
  print(k_vect[0], k_vect[1])
```
