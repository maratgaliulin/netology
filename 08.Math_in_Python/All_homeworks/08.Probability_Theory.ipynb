Задание 1

В магазин привезли устройства с 3-х разных предприятий.

Соотношение устройств следующее: 20% - продукция первого предприятия, 30% - продукция второго предприятия, 50% - продукция третьего предприятия; далее, 10% продукции первого предприятия высшего сорта, на втором предприятии - 5% и на третьем - 20% продукции высшего сорта.

Найти вероятность того, что случайно купленная новая продукция окажется высшего сорта.

```python
x1 = 0.2
x2 = 0.3
x3 = 0.5

high_sort_1 = 0.1
high_sort_2 = 0.05
high_sort_3 = 0.2
```

```python
high_sort_probability = (x1*high_sort_1 + x2*high_sort_2 + x3*high_sort_3) * 100

print(f'Вероятность продукции высшего сорта: {high_sort_probability}%')
```

Пусть брошены 3 уравновешенные монеты.

Рассмотрим 3 события:

A1 - монеты 1 и 2 упали одной и той же стороной;

A2 - монеты 2 и 3 упали одной и той же стороной;

A3 - монеты 1 и 3 упали одной и той же стороной.

Покажите, почему эти 3 события (A1, A2, A3) являются попарно независимыми, но не являются независимыми в совокупности.

```python
# События А и В являются независимыми если Р(АВ) = Р(А) * Р(В), и это то что мы имеем в каждом из событий А1, А2, А3 по отдельности
# Однако, если произошли любые 2 из этих 3 событий, то следовательно 3-е событие тоже произошло, поэтому эти 3 события не являются независимыми в совокупности
```

Задание 3

Изучите 3 распределения из библиотеки scipy: Нормальное, Экспоненциальное, Стьюдента. Используя функцию плотности распредления (pdf), постройте пять графиков плотностей для каждого распределения при разных параметрах. Запишите ваши наблюдения: как и на что влияют параметры у каждого распределения.

```python
# Нормальное распределение:

x_norm = np.arange(-3, 3, .001)
plt.plot(x_norm, norm.pdf(x_norm, 0, 1))
plt.show()

# В данном случае 1-й параметр pdf определяет наличие скошенности кривой распределения,
# а 2-й параметр определяет среднее квадратическое отклонение
```

```python
# Экспоненциальное распределение:

x_expon = np.arange(0, 10, .001)

plt.plot(x_expon, expon.pdf(x_expon, 0,1))
plt.show()

# В данном случае 1-й параметр указывает на точку отсчета "экспоненциальной" порции графика, а 2-й параметр влияет на степень вогнутости получившейся кривой
```

```python
# Распределение Стьюдента:

df = 2.74

x_stud = np.linspace(t.ppf(0.01, df),
                t.ppf(0.99, df), 100)

plt.plot(x_stud, t.pdf(x_stud, df))
plt.show()
```
