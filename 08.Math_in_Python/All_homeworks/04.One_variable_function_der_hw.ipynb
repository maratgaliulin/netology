Задание 2

Постройте график первой и второй производной функции: f(x) = sin^5(2x + 1)

на отрезке [-5; 5]

```python
# 1. обозначил переменную х2 с соответствующим отрезком от -5 до +5

x2 = np.arange(-5, 5, .05)

# 2. нашел при помощи библиотеки sympy первую и вторую производные от исходной функции 

y2 = sym.sin(2 * x + 1)**5

yp = y2.diff(x)

yp
```

```python
yp_2 = yp.diff(x)

yp_2
```

```python
# 3. сделал соответствующие python-функции для отображения через pyplot:

def primary_func(x):
  return np.sin(2 * x + 1)**5

def first_deriv(x):
  return 10 * np.sin(2*x + 1)**4 * np.cos(2*x + 1)

def second_deriv(x):
  return -20 * np.sin(2*x + 1)**5 + 80 * np.sin(2*x + 1)**3 * np.cos(2*x + 1)**2

plt.figure(figsize = (12, 12))
plt.plot(x2, primary_func(x2), label='Основная функция')
plt.plot(x2, first_deriv(x2), label='Первая производная')
plt.plot(x2, second_deriv(x2), label='Вторая производная')
plt.legend()
plt.show()
```

```python
# Основная функция и ее первая производная для наглядности:

plt.figure(figsize = (12, 5))
plt.plot(x2, primary_func(x2), label='Основная функция')
plt.plot(x2, first_deriv(x2), label='Первая производная')
plt.legend()
plt.show()
```
