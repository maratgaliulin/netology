```python
import json
```

Переведите содержимое файла purchase_log.txt в словарь purchases вида:
{‘1840e0b9d4’: ‘Продукты’, ...}

Для каждого user_id в файле visit_log.csv определите третий столбец с категорией покупки (если покупка была, сам файл visit_log.csv изменять не надо). Запишите в файлfunnel.csv визиты из файла visit_log.csv, в которых были покупки с указанием категории.

```python
purchases_raw = []  # Создаю пустой список для отгрузки информации из файла с помощью readlines()
purchases = {}    # Создаю пустой словарь
```

```python
with open('purchase_log.txt', 'r', encoding='utf-8') as f:
    purchases_raw = f.readlines()                                                   # использую readlines()
    for line in purchases_raw:
            purchases[json.loads(line)['user_id']] = json.loads(line)['category']   # Заполняю словарь парами ключ (user_id) - значение (category)
```

```python
with open('purchases.txt', 'w') as fw:
    fw.write(json.dumps(purchases))                            # Сохраняю словарь в файл purchases.txt (не знаю нужно ли было создавать этот файл, по заданию вроде бы не нужно, но сделал на всякий случай)
```

```python
# print(purchases)

i=0
for key in purchases:
  print(key, purchases[key])   # Проверка
  i+=1
  if i > 10:
    break
```

```python
with open('funnel.csv', 'w') as write_funnel:
    with open('visit_log.csv', 'r') as visits_log_list:
        for vl in visits_log_list:
            for key in purchases:
                if key in vl:
                    write_funnel.write(vl.strip() + ',' + purchases[key] + '\n')    # Заполняю файл funnel.csv
                    # print(vl.strip() + ',' + purchases[key] + '\n')         # Проверка правильности работы итератора (работает правильно)
        print('Итератор завершил работу, файл funnel.csv сформирован')
```
